```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

```

```python
# Example dataset
dataset = [['Milk', 'Eggs', 'Bread'],
           ['Milk', 'Diapers'],
           ['Milk', 'Beer', 'Bread', 'Diapers'],
           ['Eggs', 'Beer', 'Diapers'],
           ['Eggs', 'Bread', 'Beer', 'Diapers']]

```

```python
# Initialize transaction encoder
te = TransactionEncoder()
te_ary = te.fit_transform(dataset)

```

```python
# Convert encoded transaction array to dataframe
df = pd.DataFrame(te_ary, columns=te.columns_)

```

```python
# Count frequency of items
item_frequency = df.sum().sort_values(ascending=False)

```

```python
# Print item frequency
for item, frequency in item_frequency.items():
    print("Item:", item)
    print("Frequency:", frequency)
    print()
```
